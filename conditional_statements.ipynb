````python
# write a program to find the given number is positive or negative or zero?
n= int(input("please enter number here:"))
if(n>0):
   print(n,"is positive number")
elif(n<0):
    print(n,"is a negative number")
else:
    print(n,"is zero")

````

````python
"""

write aprogram to find thr grades based on the studenr marks
1. if marks in between 80 to 100 - A Grade
2. if marks in between 60 to 79 - B Grade 
3. if marks in between 40 to 59 - C Grade
4. if marks in between 25 to 39 - D Grade
5. if marks in between 0 to 24 - Fail
6. if other than above marks output is - please enter correct marks
"""
marks= int(input("enter student marks here:"))
if(marks>=80 and marks<=100):
   print(marks,"A Grade")
elif(marks>60 and marks<=79):
     print(marks,"B Grade")
elif(marks>40 and marks<=59):
     print(marks, "C Grade")
elif(marks>25 and marks<=39):
     print(marks, "D Grade")
elif(marks>0 and marks<=24):
    print(marks, "is fail")
else:
   print("please enter correct marks")
````
